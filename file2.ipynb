```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a= 1
m= 1
hbar= 1
N= 100
x= np.linspace(0, a, N+1)
dx= x[1]-x[0]
#print(x)
v0= 10
v= 0*x
print(v)
for i in range(len(x)):
    if x[i] > 3*a/4:
        v[i]= v0
print(v)        
```

```python
H = (hbar**2/(m*dx**2))*np.diag(np.ones(N-1))+v[1:-1]*np.diag(np.ones(N-1))
H= H+ (-hbar**2/(2*m*dx**2))*np.diag(np.ones(N-2), 1)
H= H+ (-hbar**2/(2*m*dx**2))*np.diag(np.ones(N-2), -1)
print(H)
```

```python
E, psi= np.linalg.eigh(H)
print(psi.T[0])
A= np.sum(psi.T[0]**2)*dx
print(A)
psi.T[0]= psi.T[0]/np.sqrt(A)
A= np.sum(psi.T[1]**2)*dx
psi.T[1]= psi.T[1]/np.sqrt(A)
A= np.sum(psi.T[3]**2)*dx
psi.T[3]= psi.T[3]/np.sqrt(A)
plt.plot(x[1:-1], psi.T[0]**2)
plt.plot(x[1:-1], psi.T[1]**2)
plt.plot(x[1:-1], psi.T[3]**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
m = 1
hbar = 1
xmin =-6.5
xmax = 6.5
N = 1000
x = np.linspace(xmin,xmax,N+1)
dx = x[1]-x[0]
p = 40
V0 = p**2/(2*m)
sig = 0.15
x0 = 2
V = 0*x
for i in range(len(V)):
    if x[i]>0 and x[i]<0.5:
        V[i]=V0
#plt.plot(x,V)

Psi0 = np.exp( -(x[1:-1]+x0)**2/sig**2)*np.exp(1j*p*(x[1:-1]+x0))
A = np.sum(np.abs(Psi0)**2*dx)
Psi0 = Psi0/np.sqrt(A)
plt.plot(x[1:-1],np.abs(Psi0))
```

```python
H = (hbar**2/(m*dx**2))*np.diag(np.ones(N-1))+V[1:-1]*np.diag(np.ones(N-1))+ (-hbar**2/(2*m*dx**2))*np.diag(np.ones(N-2),1)+ (-hbar**2/(2*m*dx**2))*np.diag(np.ones(N-2),-1)
print(H.shape)

E,psi = np.linalg.eigh(H)
psi = psi.T

A = np.sum(np.abs(psi[0])**2*dx)
psi = psi/np.sqrt(A)
plt.plot(x[1:-1],psi[0])
plt.plot(x[1:-1],psi[1])
plt.plot(x[1:-1],psi[2])
```

```python
plt.plot(x, 0.003*V)
c= 0*Psi0
for i in range(len(c)):
    c[i]= np.sum(np.conj(psi[i])*Psi0*dx)
#print(c[0], c[1], c[2])

t= 0
dt= 0.001
Psi= 0*Psi0[0]
for i in range(len(c)):
    Psi= Psi+ c[i]*psi[i]*np.exp(-1j*E[i]*t/hbar)      

plt.plot(x[1:-1], np.abs(Psi)**2)    
plt.show()
t= t+ 15*dt

Psi= 0*Psi0[0]
for i in range(len(c)):
    Psi= Psi+ c[i]*psi[i]*np.exp(-1j*E[i]*t/hbar)      

plt.plot(x[1:-1], np.abs(Psi)**2)    
plt.show()

t= t+ 55*dt

Psi= 0*Psi0[0]
for i in range(len(c)):
    Psi= Psi+ c[i]*psi[i]*np.exp(-1j*E[i]*t/hbar)      

plt.plot(x[1:-1], np.abs(Psi)**2)    
plt.show()
```
